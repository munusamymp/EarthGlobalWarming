```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import base64

filename = '/content/drive/MyDrive/1Coding Dojo/Week3/GlobalTemperatureAnomalies.csv'
df = pd.read_csv(filename)
df
```

```python
GlobalHemi = df['Hemisphere'] == 'Global'
NorthHemi = df['Hemisphere'] == 'Northern'
SouthHemi = df['Hemisphere'] == 'Southern'
dg = df.loc[GlobalHemi, :]
dn = df.loc[NorthHemi, :]
ds = df.loc[SouthHemi, :]
```

```python
dg = dg.drop(columns=['J-D',	'D-N', 'Jan', 'Feb', 'Mar',   'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python

dg.groupby('Year').sum()
x = list(dg['Year'])
y = list(dg.groupby('Year').sum())
color = {'DJF': 'k', 'MAM': 'b', 'JJA': 'g', 'SON': 'r'}
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,8));

# subplot 1:
for month in y:
  axes[0].plot(x, dg[month], c=color[month], label=month);
axes[0].set_xlim(left=1880, right=2020);
axes[0].set_ylim(bottom=-2, top=2);
axes[0].set_xlabel('Year', fontsize=16);
axes[0].set_ylabel('Temperature in Fahrenheit', fontsize=16);
axes[0].set_title('Global Hemisphere', fontsize = 18);
axes[0].legend();

# # subplot 2:
for month in y:
  axes[1].plot(x, dn[month], c=color[month], label=month);
axes[1].set_xlim(left=1880, right=2020);
axes[1].set_ylim(bottom=-2, top=2);
axes[1].set_xlabel('Year', fontsize=16);
axes[1].set_ylabel('Temperature in Fahrenheit', fontsize=16);
axes[1].set_title('Northern Hemisphere', fontsize = 18);
axes[1].legend();

# # subplot 3:
for month in y:
  axes[2].plot(x, ds[month], c=color[month], label=month);
axes[2].set_xlim(left=1880, right=2020);
axes[2].set_ylim(bottom=-2, top=2);
axes[2].set_xlabel('Year', fontsize=16);
axes[2].set_ylabel('Temperature in Fahrenheit', fontsize=16);
axes[2].set_title('Southern Hemisphere', fontsize = 18);
axes[2].legend();
```
